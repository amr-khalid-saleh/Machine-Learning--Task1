```python

from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt

# Data: study hours (X) and grades (y)
X = np.array([[3], [4], [6], [2], [5]])#2D array
y = np.array([50, 60, 70, 75, 80])#1D array

# Create and train model
model = LinearRegression()
model.fit(X, y)

# Predict for 6 hours
X_new = np.array([[6]])
y_pred = model.predict(X_new)
print(f"Predicted grade for 6 hours: {y_pred[0]:.2f}")

# Plot data and line
plt.scatter(X, y, color='blue', label='Data')
plt.plot(X, model.predict(X), color='red', label='Fit Line')
plt.xlabel('Study Hours')
plt.ylabel('Grades')
plt.title('Linear Regression')
plt.legend()
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
import numpy as np
import matplotlib.pyplot as plt

# Data: house size in sqm (X) and sold/not sold (y: 1=sold, 0=not sold)
X = np.array([[100], [20], [60], [100], [130]])
y = np.array([0, 1, 0, 1, 1])

# Create and train model
model = LogisticRegression()
model.fit(X, y)

# Predict for 100 sqm
X_new = np.array([[100]])
y_pred = model.predict(X_new)
print(f"Prediction for 100 sqm: {'Sold' if y_pred[0] == 1 else 'Not Sold'}")

# Plot data and probability curve
X_plot = np.linspace(40, 140, 100).reshape(-1, 1)
y_prob = model.predict_proba(X_plot)[:, 1]
plt.scatter(X, y, color='green', label='Data')
plt.plot(X_plot, y_prob, color='orange', label='Sale Probability')
plt.xlabel('House Size (sqm)')
plt.ylabel('Sale Probability')
plt.title('Logistic Regression')
plt.legend()
plt.show()
```
